```python
# conda activate jupyter
# jupyter notebook Run.ipynb

from openai import OpenAI

def Query(txt):
    client = OpenAI(
        base_url="http://127.0.0.1:8080/v1", timeout=86400, max_retries=0, api_key="a"
    )
    
    stream = client.chat.completions.create(
        model="deepseek-v4-pro",
        messages=[
            {
                "role": "user",
                "content": txt,
            },
        ],
        max_tokens=4096,
        stream=True,
    )    

    r = ""
    for chunk in stream:
        delta = chunk.choices[0].delta.content
        if delta:
            r = r + delta
            if "<|im_end|>" in r:
                return
            print(delta, end="", flush=True)
```

```python
Query("三越事件について教えてください。社長はなぜ解任されたのでしょうか。竹久みち、古代ペルシア秘宝展との関係を教えてください。")
```
