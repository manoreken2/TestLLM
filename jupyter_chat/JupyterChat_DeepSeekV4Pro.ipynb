```python
# see README_DeepSeekV4Pro.md to run this program


def display_markdown_txt(txt):
    from IPython.display import display, Markdown
    from IPython.display import clear_output

    clear_output(wait=True)
    display(Markdown(txt))


def Query(txt):
    from openai import OpenAI

    client = OpenAI(
        base_url="http://127.0.0.1:8080/v1",
        max_retries=0,
        api_key="a",
    )

    stream = client.chat.completions.create(
        model="deepseek",
        messages=[
            {
                "role": "user",
                "content": txt,
            },
        ],
        max_tokens=4096,
        stream=True,
    )

    r = ""
    for chunk in stream:
        delta = chunk.choices[0].delta.content

        if delta:
            r = r + delta
            if "<|im_end|" == r:
                print("Skipped <|im_end", flush=True)
                r = ""
                # 2回目実行時に発生。
                continue

            if "<|im_end|" in r:
                print("\nReached <|im_end", flush=True)
                # 処理を終了します。
                return
            if "</think" in r:
                print("\nReached <|im_end", flush=True)
                # 処理を終了します。
                return
            
            display_markdown_txt(r)
```

```python
Query("魏の司馬師の事績について教えて下さい。")
```
